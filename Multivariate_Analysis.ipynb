```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("C:/Users/Ahmet/Desktop/data/StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(3,3,1)
sns.boxplot(x="race/ethnicity",y="math score",hue="race/ethnicity",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(3,3,2)
sns.boxplot(x="race/ethnicity",y="reading score",hue="race/ethnicity",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(3,3,3)
sns.boxplot(x="race/ethnicity",y="writing score",hue="race/ethnicity",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));

plt.subplot(3,3,4)
sns.boxplot(x="gender",y="math score",hue="gender",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(3,3,5)
sns.boxplot(x="gender",y="reading score",hue="gender",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(3,3,6)
sns.boxplot(x="gender",y="writing score",hue="gender",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));

plt.subplot(3,3,7)
sns.boxplot(x="parental level of education",y="math score",hue="parental level of education",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2))
plt.xticks(rotation="75");
plt.subplot(3,3,8)
sns.boxplot(x="parental level of education",y="reading score",hue="parental level of education",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2))
plt.xticks(rotation="75");
plt.subplot(3,3,9)
sns.boxplot(x="parental level of education",y="writing score",hue="parental level of education",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2))
plt.xticks(rotation="75");
```

- We can say that the origin has a relationship with the success of the exam. When we look at the box plots in the three exams, we see that it is the same for all of them. For example, group E always has the best mean value relative to the mean.

- When we look at the effect of gender on the exam, we see that women do better in mathematics, while men seem to do better in the other two exams.

- When we come to the educational situation, we see that those with a master's degree in their family have the highest grades. Again, we can make a comment that the higher the education level the higher grades.


```python
plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
sns.boxplot(x="lunch",y="math score",hue="lunch",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(2,3,2)
sns.boxplot(x="lunch",y="reading score",hue="lunch",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(2,3,3)
sns.boxplot(x="lunch",y="writing score",hue="lunch",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));

plt.figure(figsize=(20,10))
plt.subplot(2,3,4)
sns.boxplot(x="test preparation course",y="math score",hue="test preparation course",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(2,3,5)
sns.boxplot(x="test preparation course",y="reading score",hue="test preparation course",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
plt.subplot(2,3,6)
sns.boxplot(x="test preparation course",y="writing score",hue="test preparation course",data=df,palette="deep")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, -0.2));
```

- We can easily say that those who are better fed have better grades in three exams.
- Likewise, those who attend the preparatory course have better exam grades than those who do not.
